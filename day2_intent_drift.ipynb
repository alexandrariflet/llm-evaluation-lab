# Day 2 — Intent Drift Decomposition

## Goal
Move from *measuring drift* → *understanding drift*.

We decompose multilingual drift into:
- Input similarity (are prompts equivalent?)
- Output similarity (does the model behave consistently?)
- Input–Output alignment (does the model preserve intent?)

This allows us to explain *why* drift occurs.
